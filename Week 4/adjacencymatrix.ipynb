```python

def adjm(edges,nodes):
    import numpy as np
    a = np.zeros(shape=(nodes, nodes)) #empty matrix where every value is zero, i.e. not reachable
    
    # making one where they're reachable
    for (i,j) in edges:
        a[i][j] = 1

    return (a)

```

```python
edges = [[0,1], [0,4], [1,2], [2,0], [3,4],
         [3,6], [4,0], [4,3], [4,7], [5,3],
         [5,7], [6,5], [7,4], [7,8], [8,5],
         [8,9], [9,8]]
```

```python
print(adjm(edges, 10))
```
